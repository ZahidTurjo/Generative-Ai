# Model

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=GOOGLE_API_KEY
)
```

# PromptTemplate

## Part-1

```python
from langchain_core.prompts import PromptTemplate
```

```python
template=PromptTemplate.from_template(
    """
    Explain {topic} in bangla Language
    """
) #here topic is a variable it will change later
```

```python
prompt=template.invoke(
    {"topic":"Machine Learning"}
)
print(prompt)
```

```python
response=llm.invoke(prompt)
print(response.content)
```

```python
prompt=template.invoke(
    {"topic":"NLP"}
)
print(prompt)
```

```python
response=llm.invoke(prompt)
print(response.content)
```

# local model (Hugging Face)

```python
from transformers import pipeline
pipe=pipeline(
    task="text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    device_map="auto",
    max_new_tokens=512,
    repetition_penalty=1.1
)
```

```python
from langchain_huggingface import(
    HuggingFacePipeline,
    ChatHuggingFace
)
```

```python
llm_local=HuggingFacePipeline(
    pipeline=pipe
)
chat=ChatHuggingFace(
    llm=llm_local
)
```

```python
prompt
```

```python
response=chat.invoke(prompt)
print(response.content)
```

## part-2 multiple variables in promptTemplate

```python
template = PromptTemplate.from_template(
    """Explain {topic} for a {age} years old student
    """
)
```

```python
prompt=template.invoke(
    {
        "topic":"Machine Learning",
        "age":12
    }
)
print(prompt)
```
