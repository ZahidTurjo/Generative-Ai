```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
%ls
```

```python
laoder=PyPDFLoader("1706.03762v7.pdf")
```

```python
docs=laoder.load()
len(docs)
```

```python
type(docs[0])
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=20)
splits=text_splitter.split_documents(docs)
```

```python
print(docs[0].page_content)
```

```python
len(splits)
```

```python
print(splits[0].page_content)
```

```python
embedding_model=HuggingFaceBgeEmbeddings(model_name="BAAI/bge-small-en-v1.5")
```

```python
vectorstore=Chroma.from_documents(documents=splits,embedding=embedding_model)
```

```python
retreiver=vectorstore.as_retriever()
```

```python
response=retreiver.invoke("What is self attention")
```

```python
print(response[0].page_content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

# Define a custom prompt template
prompt_template = """
You are a friendly and knowledgeable AI assistant question answering.
Use ONLY the provided context to answer.If the answer cannot be found, respond with .I don't have enough information to answer that.

Context:
{context}

Question:
{question}

Please provide a clear, detailed, and accurate answer using only the given context.
"""

prompt = ChatPromptTemplate.from_template(prompt_template)
```

```python
def format_docs(docs):
  return "\n\n".join(doc.page_content for doc in docs)
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model='llama-3.3-70b-versatile')
```

```python
rag_chain=(
    {
        "context":retreiver|
        format_docs,
        "question":
        RunnablePassthrough()
    }
    |prompt
    |llm
    |StrOutputParser()
)
```

```python
response=rag_chain.invoke("what is self attention?")
```
