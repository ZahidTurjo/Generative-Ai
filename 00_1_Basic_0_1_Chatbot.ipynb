````python
from transformers import pipeline
from langchain_huggingface import (
    HuggingFacePipeline,
    ChatHuggingFace
)

from langchain_core.prompts import(
    ChatPromptTemplate,
    MessagesPlaceholder
)
from langchain_core.messages import (
    HumanMessage,
    AIMessage,
    SystemMessage
)
from langchain_core.output_parsers import StrOutputParser
````

````python
pipe=pipeline(
    task="text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",

    max_new_tokens=128,

    temperature=0.7,

    do_sample=True,

    repetition_penalty=1.1,

    return_full_text=False,

    device_map="auto"
)
````

````python
llm=HuggingFacePipeline(pipeline=pipe)
chat=ChatHuggingFace(
    llm=llm,
)
````

````python
print("Choose AI Mode")

print("1. Angry")
print("2. Funny")
print("3. Sad")

choice = int(input("Choice : "))
if choice == 1:
    personality = "You are an angry AI assistant."

elif choice == 2:
    personality = "You are a funny AI assistant."

elif choice == 3:
    personality = "You are a sad AI assistant."

else:
    personality = "You are a helpful AI assistant."
print(choice)
print(personality)
````

````python
prompt=ChatPromptTemplate(
    [
        ("system",personality),
        MessagesPlaceholder("history"),
        ("human","{input}")
    ]
)
````

System Prompt

↓

আগের Conversation

↓

বর্তমান User Question

````python
parser=StrOutputParser() #এটা AI-এর output-কে string বানিয়ে দেয়।
````

LangChain Expression Language (LCEL)

````python
chain=prompt | chat | parser
````



```
Prompt
   │
   ▼
Chat Model
   │
   ▼
String Output
```



---



````python
history=[]

print("Type 0 to exit\n")

while True:

  user_input=input("You : ")
  if user_input=="0":
    break
  response=chain.invoke(
      {
          "history":history,
          "input":user_input
      }
  )


  print("Bot : ",response)

  history.append(
      AIMessage(content=response)
  )

  if len(history)>20:
    # print(len(history))
    history=history[-20:]
````

## add HumanMessage in the chat

````python
history=[]

print("Type 0 to exit\n")

while True:

  user_input=input("You : ")
  if user_input=="0":
    break
  history.append(
      HumanMessage(content=user_input)
  )
  response=chain.invoke(
      {
          "history":history,
          "input":user_input
      }
  )


  print("Bot : ",response)

  history.append(
      AIMessage(content=response)
  )

  if len(history)>10:
    # print(len(history))
    history=history[-10:]
````
