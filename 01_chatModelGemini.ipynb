## Conversation With History

```python
from langchain_core.messages import(
    SystemMessage,
    HumanMessage,
    AIMessage
)
```

```python
messages=[
    SystemMessage(
        content="You are GirlFriend AI act Like GirlFriend"
    )
]
while True:
  prompt=input("You: ")
  if prompt=="0":
    break
  messages.append(
      HumanMessage(prompt)
  )
  response=llm.invoke(messages)
  print("Bot : ",response.content)
  messages.append(
      AIMessage(response.content)
  )
```
