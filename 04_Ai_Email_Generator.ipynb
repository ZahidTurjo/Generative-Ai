```python
from langchain_groq import ChatGroq
from langchain_core.prompts import (
    ChatPromptTemplate,
    FewShotChatMessagePromptTemplate)
```

```python
from getpass import getpass
import os
os.environ['GROQ_API_KEY']=getpass("Enter Groq API key: ")
```

# MOdel

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

# Few Shot Examples

```python
examples = [

{
    "name":"Rahim",
    "company":"Google",
    "position":"Machine Learning Engineer",
    "tone":"Professional",

    "email":"""

Subject: Application for Machine Learning Engineer

Dear Hiring Manager,

My name is Rahim and I would like to express my interest in the Machine Learning Engineer position at Google.

I believe my experience in AI and machine learning makes me a strong candidate.

Thank you for your consideration.

Best Regards,
Rahim
"""
},

{
    "name":"Karim",
    "company":"Microsoft",
    "position":"Data Scientist",
    "tone":"Friendly",

    "email":"""

Subject: Excited to Join Microsoft

Hello,

I hope you are doing well.

My name is Karim and I am excited to apply for the Data Scientist position at Microsoft.

I would love the opportunity to contribute to your amazing team.

Looking forward to hearing from you.

Best,
Karim
"""
}

]
```

## Example Prompt

```python
example_prompt=ChatPromptTemplate.from_messages(
    [
        ("human",
         """
         Name:{name}
         Company:{company}
         Position:{position}
         Tone:{tone}
         """
         ),
        (
            ("ai","{email}")
        )
    ]
)
```

## Few shot prompt

```python
few_shot_prompt=FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples
)
```

## Partial Prompt

```python
system=ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """
            You are an expert HR email writer.

            Write professional emails.

            Today's role is:
            {role}
            """
        ),
        few_shot_prompt,
        (
            "human",
            """
            Name :{name}
            Company :{company}
            Position :{position}
            Tone :{tone}
            Generate only the email
            """
        )

    ]
)
```

```python
prompt=system.partial(
    role="AI Email Generator"
)
```

```python
final_prompt=prompt.invoke(
    {
        "name":"Zahid",
        "company":"OpenAI",
        "position":"AI Engineer",
        "tone":"Professional"
    }
)
```

```python
final_prompt
```

```python
response=llm.invoke(final_prompt)
print(response.content)
```
