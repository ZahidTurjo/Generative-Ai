```python
import os
from getpass import getpass
os.environ['GROQ_API_KEY'] = getpass('GET Groq Api key:')
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import(
    ChatPromptTemplate,
    FewShotChatMessagePromptTemplate
)
```

## MOdel

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.5
)
```

## Few Shot Example

```python
examples = [

{
    "role":"Python Developer",
    "experience":"Fresher",

    "output":"""
Q1. What is a Python list?

Answer:
A list is an ordered, mutable collection.

Q2. Difference between tuple and list?

Answer:
Lists are mutable whereas tuples are immutable.
"""
},

{
    "role":"Data Analyst",
    "experience":"Intermediate",

    "output":"""
Q1. What is VLOOKUP?

Answer:
VLOOKUP searches for a value in a table.

Q2. What is a Pivot Table?

Answer:
A Pivot Table summarizes data quickly.
"""
}

]
```

## example Prompt

```python
example_prompt=ChatPromptTemplate.from_messages(
    [
        ("human",
         """
         Role : {role}
         Experience:{experience}
         """),
        (
            "ai","{output}"
        )
    ]
)
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(

examples=examples,

example_prompt=example_prompt

)
```

```python
few_shot_prompt
```

## Partical Prompt

```python
prompt=ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """
            You are an expert {interviewer}
            Generate interview questions with answers
            """
        ),
        few_shot_prompt,
        (
            "human",
            """
            Role : {role}
            Experience:{experience}
            Difficulty:{difficulty}
            Generate {number} interview questions with answers
            """
        )

    ]
)
```

```python
prompt=prompt.partial(
    interviewer="Technical Interviewer"
)
```

```python
prompt
```

```python
final_prompt=prompt.invoke(
    {
        "role":"ML",
        "experience":"Fresher",
        "difficulty":"Easy",
        "number":5
    }
)
response=llm.invoke(final_prompt)
print(response.content)
```

```python
final_prompt=prompt.invoke(
    {
        "role":"Supply Chain Analyst",
        "experience":"Fresher",
        "difficulty":"hard",
        "number":6
    }
)
response=llm.invoke(final_prompt)
print(response.content)
```
