# CharacterSplitter

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text = """
Machine Learning is a branch of Artificial Intelligence.

Deep Learning is a subset of Machine Learning.

Neural Networks consist of multiple layers.

Regression predicts continuous values.

Classification predicts categorical labels.

Decision Trees are supervised learning algorithms.

Random Forest is an ensemble technique.

Support Vector Machine separates classes.

KNN is a lazy learner.

Naive Bayes is based on probability.
"""
```

```python
splitter=CharacterTextSplitter(
    separator="\n",
    chunk_size=100,
    chunk_overlap=0
)
```

Split where the new Line started

```python
chunks=splitter.split_text(text)
```

```python
len(chunks)
```

```python
for i,chunk in enumerate(chunks):
  print("="*50)
  print(f"Chunk {i+1}")
  print(chunk)
```

# RecursiveCharacterTextSplitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = """
Machine Learning is a branch of Artificial Intelligence.

Deep Learning is a subset of Machine Learning.

Neural Networks are inspired by the human brain.

Regression predicts continuous values.

Classification predicts labels.
"""
```

```python
splitter=RecursiveCharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=0
)
```

```python
chunks=splitter.split_text(text)
for i,chunk in enumerate(chunks):
  print("="*50)
  print(f"Chunk {i+1}")
  print(chunk)
```



---
| Use Case        | Chunk Size |
| --------------- | ---------: |
| Short Notes     |    300–500 |
| Articles        |    500–800 |
| Research Papers |   800–1200 |
| Books           |  1000–1500 |
| Legal Documents |  1200–2000 |



# Overlap

```python
splitter=RecursiveCharacterTextSplitter(
    chunk_size=20,
    chunk_overlap=5
)
chunks=splitter.split_text(text)
for i,chunk in enumerate(chunks):
  print("="*50)
  print(f"Chunks : {i+1}")
  print(chunk)
```



---
| Chunk Size | Good Overlap |
| ---------- | -----------: |
| 300        |           50 |
| 500        |       50–100 |
| 1000       |      100–200 |
| 1500       |      150–300 |



## OverLappin with all Chunks

```python
splitter=RecursiveCharacterTextSplitter(
    separators=[""],
    chunk_size=50,
    chunk_overlap=5
)
chunks=splitter.split_text(text)
for i,chunk in enumerate(chunks):
  print("="*50)
  print(f"Chunks : {i+1}")
  print(chunk)
```
